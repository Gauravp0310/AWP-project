# EDA Exploratory Data Analysis.

**Amount of Water Use in Cement**

## Count Plot

```python
plt.figure(figsize=(12,4))
sns.countplot(x=df1['Cement'],order=df1['Cement'].value_counts().index.sort_values())
plt.xlabel('Cement',fontsize=13)
plt.ylabel('Water',fontsize=13)
plt.title('Amount of Water  vs  Cement',fontsize=16)
```

# Scatter plot

```python
# scatter plot of Water and Cement
plt.figure(figsize=[12,7])
plt.scatter(y='Concrete compressive strength',x='Cement',edgecolors='yellow',data=df)
plt.ylabel('Concrete compressive strength')
plt.xlabel('Cement')
```

# Box Plot:
**we plot the outlier that is present inside the dataset:**

```python
l=['Cement','Blast Furnace Slag','Fly Ash','Water','Superplasticizer','Coarse Aggregate','Fine Aggregate','Age','Concrete compressive strength']
for i in l:
    sns.boxplot(x=df[i])
    plt.show()
```
